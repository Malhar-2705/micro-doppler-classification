```python
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras import layers, models, optimizers, regularizers, Model
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, roc_curve, auc, precision_recall_fscore_support, accuracy_score
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import time
import pandas as pd
from datetime import datetime

# Install LIME if not available
try:
    import lime
    from lime import lime_image
    from lime.wrappers.scikit_image import SegmentationAlgorithm
except ImportError:
    print("Installing LIME...")
    import subprocess
    subprocess.check_call(["pip", "install", "lime"])
    import lime
    from lime import lime_image
    from lime.wrappers.scikit_image import SegmentationAlgorithm

from skimage.segmentation import mark_boundaries

# Category labels
categories = {
    '3_long_blade_rotor': 0,
    '3_short_blade_rotor': 1,
    'Bird': 2,
    'Bird+mini-helicopter': 3,
    'drone': 4,
    'rc_plane': 5
}

# Time tracking class
class TimeTracker:
    def __init__(self):
        self.times = {}
        self.start_times = {}
    
    def start(self, operation):
        self.start_times[operation] = time.time()
    
    def end(self, operation):
        if operation in self.start_times:
            elapsed = time.time() - self.start_times[operation]
            if operation not in self.times:
                self.times[operation] = []
            self.times[operation].append(elapsed)
            return elapsed
        return None
    
    def get_summary(self):
        summary = {}
        for op, times in self.times.items():
            summary[op] = {
                'total_time': sum(times),
                'avg_time': np.mean(times),
                'min_time': min(times),
                'max_time': max(times),
                'count': len(times)
            }
        return summary

# Initialize time tracker
timer = TimeTracker()

# Load images and labels
def load_images_from_folder(folder):
    timer.start('data_loading')
    images, labels = [], []
    for category, label in categories.items():
        category_folder = os.path.join(folder, category)
        if not os.path.exists(category_folder):
            print(f"Warning: Category folder '{category_folder}' does not exist.")
            continue
        
        for filename in os.listdir(category_folder):
            if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
                try:
                    img = load_img(os.path.join(category_folder, filename), target_size=(224, 224))
                    images.append(img_to_array(img))
                    labels.append(label)
                except Exception as e:
                    print(f"Error loading image {filename}: {e}")
    
    timer.end('data_loading')
    return np.array(images), np.array(labels)

# Dataset path
dataset_path = r"/kaggle/input/micro-doppler-dataset/new1"
X, y = load_images_from_folder(dataset_path)

# Normalize and one-hot encode
timer.start('data_preprocessing')
X = X.astype('float32') / 255.0
y = to_categorical(y, num_classes=6)
timer.end('data_preprocessing')

# Train-validation split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Model creation function for ablation study
def create_model(skip_connections=True, dropout_rate=0.4, dense_units=256, num_dense_layers=3):
    timer.start('model_creation')
    
    base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    base_model.trainable = False
    
    x = base_model.output
    x = layers.GlobalAveragePooling2D()(x)
    
    if num_dense_layers >= 1:
        dense1 = layers.Dense(dense_units, activation='relu', kernel_regularizer=regularizers.l2(0.001))(x)
        if dropout_rate > 0:
            dropout1 = layers.Dropout(dropout_rate)(dense1)
        else:
            dropout1 = dense1
        
        if num_dense_layers >= 2:
            dense2 = layers.Dense(dense_units, activation='relu', kernel_regularizer=regularizers.l2(0.001))(dropout1)
            
            if skip_connections and num_dense_layers >= 2:
                skip1 = layers.Add()([dense1, dense2])
            else:
                skip1 = dense2
            
            if dropout_rate > 0:
                dropout2 = layers.Dropout(dropout_rate)(skip1)
            else:
                dropout2 = skip1
            
            if num_dense_layers >= 3:
                dense3 = layers.Dense(dense_units, activation='relu', kernel_regularizer=regularizers.l2(0.001))(dropout2)
                
                if skip_connections:
                    skip2 = layers.Add()([skip1, dense3])
                else:
                    skip2 = dense3
                
                if dropout_rate > 0:
                    dropout3 = layers.Dropout(dropout_rate)(skip2)
                else:
                    dropout3 = skip2
                
                output = layers.Dense(6, activation='softmax')(dropout3)
            else:
                output = layers.Dense(6, activation='softmax')(dropout2)
        else:
            output = layers.Dense(6, activation='softmax')(dropout1)
    else:
        output = layers.Dense(6, activation='softmax')(x)
    
    model = Model(inputs=base_model.input, outputs=output)
    timer.end('model_creation')
    
    # Store the base model reference for later fine-tuning
    model.base_model = base_model
    return model

# Create main model
model = create_model()
model.summary()

# Compile and initial training
timer.start('initial_training')
optimizer = optimizers.Adam(learning_rate=0.0001)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_val, y_val))
timer.end('initial_training')

# Fine-tune: Unfreeze last layers of base model
timer.start('fine_tuning')
# Access the base model using the stored reference
base_model = model.base_model
base_model.trainable = True

# Freeze the first 75 layers (keep early layers frozen)
for layer in base_model.layers[:75]:
    layer.trainable = False

# Recompile and fine-tune
model.compile(optimizer=optimizers.Adam(learning_rate=0.00006), loss='categorical_crossentropy', metrics=['accuracy'])
history_finetune = model.fit(X_train, y_train, epochs=150, batch_size=32, validation_data=(X_val, y_val))
timer.end('fine_tuning')

# Evaluation
timer.start('evaluation')
loss, accuracy = model.evaluate(X_val, y_val)
print(f"Validation Accuracy: {accuracy*100:.2f}%")
timer.end('evaluation')

# Prediction and metrics
timer.start('prediction')
y_pred_probs = model.predict(X_val)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = np.argmax(y_val, axis=1)
timer.end('prediction')

# ORIGINAL VISUALIZATIONS (PRESERVED)

# Confusion matrix
cm = confusion_matrix(y_true, y_pred, normalize='true')
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='.2f', cmap='Blues',
            xticklabels=categories.keys(),
            yticklabels=categories.keys())
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Normalized Confusion Matrix")
plt.show()

# ROC curve
plt.figure(figsize=(6, 5))
for i in range(6):
    fpr, tpr, _ = roc_curve(y_val[:, i], y_pred_probs[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'Class {i} (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()

# Precision, Recall, F1-score
precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, average=None)
accuracy = accuracy_score(y_true, y_pred)

print("\nPrecision, Recall, F1-score Table:")
print("Class\t\tPrecision\tRecall\t\tF1-score")
for i, class_name in enumerate(categories.keys()):
    print(f"{class_name}\t{precision[i]:.2f}\t\t{recall[i]:.2f}\t\t{f1[i]:.2f}")
print(f"\nOverall Accuracy: {accuracy*100:.2f}%")

# Combine histories
def combine_history(h1, h2):
    history = {}
    for key in h1.history.keys():
        history[key] = h1.history[key] + h2.history[key]
    return history

full_history = combine_history(history, history_finetune)

# Accuracy and loss plots
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(full_history['accuracy'], label='Training Accuracy')
plt.plot(full_history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(full_history['loss'], label='Training Loss')
plt.plot(full_history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.tight_layout()
plt.show()

# NEW ANALYSES

# 1. TIME ANALYSIS
print("\n" + "="*50)
print("TIME ANALYSIS")
print("="*50)

time_summary = timer.get_summary()
time_df = pd.DataFrame(time_summary).T
print("\nTime Analysis Summary:")
print(time_df)

# Time analysis visualization
plt.figure(figsize=(12, 8))

# Plot 1: Total time by operation
plt.subplot(2, 2, 1)
operations = list(time_summary.keys())
total_times = [time_summary[op]['total_time'] for op in operations]
plt.bar(operations, total_times, color='skyblue')
plt.title('Total Time by Operation')
plt.xlabel('Operation')
plt.ylabel('Time (seconds)')
plt.xticks(rotation=45)

# Plot 2: Average time per operation
plt.subplot(2, 2, 2)
avg_times = [time_summary[op]['avg_time'] for op in operations]
plt.bar(operations, avg_times, color='lightcoral')
plt.title('Average Time per Operation')
plt.xlabel('Operation')
plt.ylabel('Time (seconds)')
plt.xticks(rotation=45)

# Plot 3: Inference time analysis
plt.subplot(2, 2, 3)
# Measure inference time for different batch sizes
batch_sizes = [1, 8, 16, 32, 64]
inference_times = []
sample_data = X_val[:64]  # Use first 64 samples

for batch_size in batch_sizes:
    times = []
    for i in range(0, min(64, len(sample_data)), batch_size):
        batch = sample_data[i:i+batch_size]
        start_time = time.time()
        _ = model.predict(batch, verbose=0)
        end_time = time.time()
        times.append((end_time - start_time) / len(batch))  # Time per sample
    inference_times.append(np.mean(times))

plt.plot(batch_sizes, inference_times, marker='o', color='green')
plt.title('Inference Time vs Batch Size')
plt.xlabel('Batch Size')
plt.ylabel('Time per Sample (seconds)')

# Plot 4: Training time breakdown
plt.subplot(2, 2, 4)
training_breakdown = {
    'Initial Training': time_summary.get('initial_training', {}).get('total_time', 0),
    'Fine Tuning': time_summary.get('fine_tuning', {}).get('total_time', 0),
    'Data Loading': time_summary.get('data_loading', {}).get('total_time', 0),
    'Other': sum([time_summary[op]['total_time'] for op in time_summary.keys() 
                  if op not in ['initial_training', 'fine_tuning', 'data_loading']])
}
plt.pie(training_breakdown.values(), labels=training_breakdown.keys(), autopct='%1.1f%%')
plt.title('Training Time Breakdown')

plt.tight_layout()
plt.show()

# 2. ABLATION STUDY
print("\n" + "="*50)
print("ABLATION STUDY")
print("="*50)

ablation_results = []

# Define ablation configurations
ablation_configs = [
    {'name': 'Original Model', 'skip_connections': True, 'dropout_rate': 0.4, 'dense_units': 256, 'num_dense_layers': 3},
    {'name': 'No Skip Connections', 'skip_connections': False, 'dropout_rate': 0.4, 'dense_units': 256, 'num_dense_layers': 3},
    {'name': 'No Dropout', 'skip_connections': True, 'dropout_rate': 0.0, 'dense_units': 256, 'num_dense_layers': 3},
    {'name': 'Smaller Dense Units', 'skip_connections': True, 'dropout_rate': 0.4, 'dense_units': 128, 'num_dense_layers': 3},
    {'name': 'Fewer Dense Layers', 'skip_connections': True, 'dropout_rate': 0.4, 'dense_units': 256, 'num_dense_layers': 2},
    {'name': 'High Dropout', 'skip_connections': True, 'dropout_rate': 0.6, 'dense_units': 256, 'num_dense_layers': 3}
]

for config in ablation_configs:
    print(f"\nTesting: {config['name']}")
    timer.start(f"ablation_{config['name']}")
    
    # Create model with specific configuration
    ablation_model = create_model(
        skip_connections=config['skip_connections'],
        dropout_rate=config['dropout_rate'],
        dense_units=config['dense_units'],
        num_dense_layers=config['num_dense_layers']
    )
    
    # Compile model
    ablation_model.compile(
        optimizer=optimizers.Adam(learning_rate=0.0001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    # Train for fewer epochs for ablation study
    ablation_history = ablation_model.fit(
        X_train, y_train,
        epochs=20,  # Reduced epochs for ablation study
        batch_size=32,
        validation_data=(X_val, y_val),
        verbose=0
    )
    
    # Evaluate
    val_loss, val_accuracy = ablation_model.evaluate(X_val, y_val, verbose=0)
    train_time = timer.end(f"ablation_{config['name']}")
    
    # Count parameters - FIXED METHOD
    total_params = ablation_model.count_params()
    trainable_params = sum([layer.count_params() for layer in ablation_model.layers if layer.trainable])
    
    ablation_results.append({
        'Configuration': config['name'],
        'Validation Accuracy': val_accuracy,
        'Validation Loss': val_loss,
        'Training Time (s)': train_time,
        'Total Parameters': total_params,
        'Trainable Parameters': trainable_params,
        'Final Train Acc': ablation_history.history['accuracy'][-1],
        'Final Val Acc': ablation_history.history['val_accuracy'][-1]
    })

# Ablation results
ablation_df = pd.DataFrame(ablation_results)
print("\nAblation Study Results:")
print(ablation_df.round(4))

# Ablation study visualization
plt.figure(figsize=(15, 10))

# Plot 1: Validation Accuracy Comparison
plt.subplot(2, 3, 1)
configs = ablation_df['Configuration']
val_accs = ablation_df['Validation Accuracy']
bars = plt.bar(range(len(configs)), val_accs, color='lightblue')
plt.title('Validation Accuracy by Configuration')
plt.xlabel('Configuration')
plt.ylabel('Accuracy')
plt.xticks(range(len(configs)), configs, rotation=45, ha='right')
for i, bar in enumerate(bars):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.001, 
             f'{val_accs.iloc[i]:.3f}', ha='center', va='bottom')

# Plot 2: Training Time Comparison
plt.subplot(2, 3, 2)
train_times = ablation_df['Training Time (s)']
plt.bar(range(len(configs)), train_times, color='lightcoral')
plt.title('Training Time by Configuration')
plt.xlabel('Configuration')
plt.ylabel('Time (seconds)')
plt.xticks(range(len(configs)), configs, rotation=45, ha='right')

# Plot 3: Parameters vs Accuracy
plt.subplot(2, 3, 3)
trainable_params = ablation_df['Trainable Parameters']
plt.scatter(trainable_params, val_accs, color='green', s=100)
plt.title('Trainable Parameters vs Accuracy')
plt.xlabel('Trainable Parameters')
plt.ylabel('Validation Accuracy')
for i, config in enumerate(configs):
    plt.annotate(config.split()[0], (trainable_params.iloc[i], val_accs.iloc[i]), 
                xytext=(5, 5), textcoords='offset points', fontsize=8)

# Plot 4: Accuracy vs Training Time
plt.subplot(2, 3, 4)
plt.scatter(train_times, val_accs, color='purple', s=100)
plt.title('Training Time vs Accuracy')
plt.xlabel('Training Time (seconds)')
plt.ylabel('Validation Accuracy')
for i, config in enumerate(configs):
    plt.annotate(config.split()[0], (train_times.iloc[i], val_accs.iloc[i]), 
                xytext=(5, 5), textcoords='offset points', fontsize=8)

# Plot 5: Train vs Validation Accuracy
plt.subplot(2, 3, 5)
train_accs = ablation_df['Final Train Acc']
plt.scatter(train_accs, val_accs, color='orange', s=100)
plt.plot([0.5, 1], [0.5, 1], 'r--', alpha=0.5)  # Perfect correlation line
plt.title('Training vs Validation Accuracy')
plt.xlabel('Training Accuracy')
plt.ylabel('Validation Accuracy')
for i, config in enumerate(configs):
    plt.annotate(config.split()[0], (train_accs.iloc[i], val_accs.iloc[i]), 
                xytext=(5, 5), textcoords='offset points', fontsize=8)

# Plot 6: Validation Loss Comparison
plt.subplot(2, 3, 6)
val_losses = ablation_df['Validation Loss']
plt.bar(range(len(configs)), val_losses, color='lightyellow')
plt.title('Validation Loss by Configuration')
plt.xlabel('Configuration')
plt.ylabel('Loss')
plt.xticks(range(len(configs)), configs, rotation=45, ha='right')

plt.tight_layout()
plt.show()

# 3. LIME ANALYSIS
print("\n" + "="*50)
print("LIME ANALYSIS")
print("="*50)

def model_predict_fn(images):
    """Prediction function for LIME."""
    return model.predict(images)

# Initialize LIME explainer
explainer = lime_image.LimeImageExplainer()

# Select sample images for explanation (one from each class)
sample_indices = []
for class_id in range(6):
    class_indices = np.where(y_true == class_id)[0]
    if len(class_indices) > 0:
        sample_indices.append(class_indices[0])

print(f"Analyzing {len(sample_indices)} sample images with LIME...")

# Generate LIME explanations
lime_explanations = []
for i, idx in enumerate(sample_indices):
    print(f"Generating explanation {i+1}/{len(sample_indices)}...")
    
    image = X_val[idx]
    true_class = y_true[idx]
    pred_class = y_pred[idx]
    pred_prob = y_pred_probs[idx]
    
    # Generate explanation
    explanation = explainer.explain_instance(
        image, 
        model_predict_fn,
        top_labels=6,
        hide_color=0,
        num_samples=100
    )
    
    lime_explanations.append({
        'image': image,
        'explanation': explanation,
        'true_class': true_class,
        'pred_class': pred_class,
        'pred_prob': pred_prob,
        'idx': idx
    })

# Visualize LIME explanations
fig, axes = plt.subplots(len(sample_indices), 4, figsize=(16, 4*len(sample_indices)))
if len(sample_indices) == 1:
    axes = axes.reshape(1, -1)

class_names = list(categories.keys())

for i, lime_data in enumerate(lime_explanations):
    image = lime_data['image']
    explanation = lime_data['explanation']
    true_class = lime_data['true_class']
    pred_class = lime_data['pred_class']
    pred_prob = lime_data['pred_prob']
    
    # Original image
    axes[i, 0].imshow(image)
    axes[i, 0].set_title(f'Original\nTrue: {class_names[true_class]}\nPred: {class_names[pred_class]}')
    axes[i, 0].axis('off')
    
    # LIME explanation for predicted class
    temp, mask = explanation.get_image_and_mask(
        explanation.top_labels[0], 
        positive_only=True, 
        num_features=10, 
        hide_rest=False
    )
    axes[i, 1].imshow(mark_boundaries(temp / 2 + 0.5, mask))
    axes[i, 1].set_title(f'LIME: Positive Features\n(Prob: {pred_prob[pred_class]:.3f})')
    axes[i, 1].axis('off')
    
    # LIME explanation for predicted class (negative)
    temp, mask = explanation.get_image_and_mask(
        explanation.top_labels[0], 
        positive_only=False, 
        num_features=10, 
        hide_rest=False
    )
    axes[i, 2].imshow(mark_boundaries(temp / 2 + 0.5, mask))
    axes[i, 2].set_title('LIME: All Features')
    axes[i, 2].axis('off')
    
    # Prediction probabilities
    axes[i, 3].bar(range(6), pred_prob, color='skyblue')
    axes[i, 3].set_title('Prediction Probabilities')
    axes[i, 3].set_xlabel('Class')
    axes[i, 3].set_ylabel('Probability')
    axes[i, 3].set_xticks(range(6))
    axes[i, 3].set_xticklabels([name[:10] for name in class_names], rotation=45)

plt.tight_layout()
plt.show()

# LIME feature importance analysis
print("\nLIME Feature Importance Analysis:")
feature_importance_summary = []

for i, lime_data in enumerate(lime_explanations):
    explanation = lime_data['explanation']
    true_class = lime_data['true_class']
    pred_class = lime_data['pred_class']
    
    # Get feature importance for the predicted class
    features = explanation.local_exp[explanation.top_labels[0]]
    
    # Calculate positive and negative importance
    positive_importance = sum([weight for _, weight in features if weight > 0])
    negative_importance = sum([weight for _, weight in features if weight < 0])
    total_features = len(features)
    
    feature_importance_summary.append({
        'Sample': i+1,
        'True Class': class_names[true_class],
        'Predicted Class': class_names[pred_class],
        'Positive Importance': positive_importance,
        'Negative Importance': abs(negative_importance),
        'Net Importance': positive_importance + negative_importance,
        'Total Features': total_features
    })

lime_summary_df = pd.DataFrame(feature_importance_summary)
print(lime_summary_df.round(4))

# LIME summary visualization
plt.figure(figsize=(12, 8))

# Plot 1: Positive vs Negative Importance
plt.subplot(2, 2, 1)
x = range(len(lime_summary_df))
plt.bar(x, lime_summary_df['Positive Importance'], alpha=0.7, label='Positive', color='green')
plt.bar(x, -lime_summary_df['Negative Importance'], alpha=0.7, label='Negative', color='red')
plt.title('LIME Feature Importance by Sample')
plt.xlabel('Sample')
plt.ylabel('Importance')
plt.legend()
plt.xticks(x, [f"S{i+1}" for i in x])

# Plot 2: Net Importance
plt.subplot(2, 2, 2)
plt.bar(x, lime_summary_df['Net Importance'], color='blue', alpha=0.7)
plt.title('Net LIME Importance by Sample')
plt.xlabel('Sample')
plt.ylabel('Net Importance')
plt.xticks(x, [f"S{i+1}" for i in x])

# Plot 3: Class-wise LIME analysis
plt.subplot(2, 2, 3)
class_importance = lime_summary_df.groupby('Predicted Class')['Net Importance'].mean()
plt.bar(range(len(class_importance)), class_importance.values, color='orange', alpha=0.7)
plt.title('Average LIME Importance by Predicted Class')
plt.xlabel('Class')
plt.ylabel('Average Net Importance')
plt.xticks(range(len(class_importance)), [name[:10] for name in class_importance.index], rotation=45)

# Plot 4: Prediction confidence vs LIME importance
plt.subplot(2, 2, 4)
max_probs = [lime_data['pred_prob'][lime_data['pred_class']] for lime_data in lime_explanations]
net_importance = lime_summary_df['Net Importance']
plt.scatter(max_probs, net_importance, color='purple', s=100)
plt.title('Prediction Confidence vs LIME Importance')
plt.xlabel('Prediction Confidence')
plt.ylabel('Net LIME Importance')
for i, (x, y) in enumerate(zip(max_probs, net_importance)):
    plt.annotate(f'S{i+1}', (x, y), xytext=(5, 5), textcoords='offset points')

plt.tight_layout()
plt.show()
# Final summary
print("\n" + "="*70)
print("COMPREHENSIVE ANALYSIS SUMMARY")
print("="*70)

print(f"\n📊 PERFORMANCE METRICS:")
print(f"   • Validation Accuracy: {accuracy*100:.2f}%")
print(f"   • Validation Loss: {loss:.4f}")

print(f"\n⏱️ TIME ANALYSIS:")
total_time = sum([time_summary[op]['total_time'] for op in time_summary.keys()])
print(f"   • Total Execution Time: {total_time:.2f} seconds ({total_time/60:.2f} minutes)")
print(f"   • Data Loading Time: {time_summary.get('data_loading', {}).get('total_time', 0):.2f} seconds")
print(f"   • Initial Training Time: {time_summary.get('initial_training', {}).get('total_time', 0):.2f} seconds")
print(f"   • Fine Tuning Time: {time_summary.get('fine_tuning', {}).get('total_time', 0):.2f} seconds")

print(f"\n🔬 ABLATION STUDY:")
best_config = ablation_df.loc[ablation_df['Validation Accuracy'].idxmax()]
print(f"   • Best Configuration: {best_config['Configuration']}")
print(f"   • Best Validation Accuracy: {best_config['Validation Accuracy']:.4f}")
print(f"   • Performance Drop without Skip Connections: {(ablation_df.loc[0, 'Validation Accuracy'] - ablation_df.loc[1, 'Validation Accuracy']):.4f}")
print(f"   • Performance Drop without Dropout: {(ablation_df.loc[0, 'Validation Accuracy'] - ablation_df.loc[2, 'Validation Accuracy']):.4f}")

print(f"\n🔍 LIME ANALYSIS:")
avg_positive_importance = lime_summary_df['Positive Importance'].mean()
avg_negative_importance = lime_summary_df['Negative Importance'].mean()
print(f"   • Average Positive Feature Importance: {avg_positive_importance:.4f}")
print(f"   • Average Negative Feature Importance: {avg_negative_importance:.4f}")
```
