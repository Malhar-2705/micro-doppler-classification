```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import numpy as np
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
import math
import time

# Enable anomaly detection for debugging
torch.autograd.set_detect_anomaly(True)

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Optimized data transforms with augmentation
transform_train = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.5], [0.5])
])

transform_val = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.ToTensor(),
    transforms.Normalize([0.5], [0.5])
])

# Dataset loading
train_dataset = datasets.ImageFolder("/kaggle/input/micro-doppler-dataset/new1", transform=transform_train)
val_dataset = datasets.ImageFolder("/kaggle/input/micro-doppler-dataset/new1", transform=transform_val)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=4, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=4, pin_memory=True)

num_classes = len(train_dataset.classes)
print(f"Number of classes: {num_classes}")

# Efficient Mamba-like State Space Block
class MambaBlock(nn.Module):
    def __init__(self, d_model, d_state=16, expand=2, conv_size=3):
        super().__init__()
        self.d_model = d_model
        self.d_state = d_state
        self.d_inner = int(expand * d_model)
        
        # Input projection
        self.in_proj = nn.Linear(d_model, self.d_inner * 2, bias=False)
        
        # Depthwise convolution
        self.conv1d = nn.Conv1d(
            in_channels=self.d_inner,
            out_channels=self.d_inner,
            kernel_size=conv_size,
            padding=conv_size // 2,
            groups=self.d_inner,
            bias=False
        )
        
        # State space parameters (simplified)
        self.x_proj = nn.Linear(self.d_inner, d_state, bias=False)
        self.dt_proj = nn.Linear(self.d_inner, d_state, bias=False)
        
        # Output projection
        self.out_proj = nn.Linear(self.d_inner, d_model, bias=False)
        
        # Layer norm
        self.norm = nn.LayerNorm(d_model)
        
    def forward(self, x):
        # x: (B, L, D)
        B, L, D = x.shape
        residual = x
        
        # Input projection
        xz = self.in_proj(x)  # (B, L, 2*d_inner)
        x, z = xz.chunk(2, dim=-1)  # Each: (B, L, d_inner)
        
        # Convolution
        x = x.transpose(1, 2)  # (B, d_inner, L)
        x = self.conv1d(x)
        x = x.transpose(1, 2)  # (B, L, d_inner)
        x = F.silu(x)
        
        # Simplified state space operation
        dt = self.dt_proj(x)  # (B, L, d_state)
        x_state = self.x_proj(x)  # (B, L, d_state)
        
        # Apply state space transformation (simplified)
        x_state = torch.cumsum(x_state * torch.sigmoid(dt), dim=1)
        
        # Combine with gate
        x = x * torch.sigmoid(z)
        
        # Output projection
        x = self.out_proj(x)
        
        # Residual connection and norm
        return self.norm(residual + x)

# Lightweight Vision Mamba
class CompactVisionMamba(nn.Module):
    def __init__(self, img_size=64, patch_size=8, in_chans=3, embed_dim=96, 
                 depth=6, d_state=16, expand=2):
        super().__init__()
        self.patch_size = patch_size
        self.num_patches = (img_size // patch_size) ** 2
        self.embed_dim = embed_dim
        
        # Patch embedding with depthwise separable conv
        self.patch_embed = nn.Sequential(
            nn.Conv2d(in_chans, embed_dim // 2, 3, 1, 1, bias=False),
            nn.BatchNorm2d(embed_dim // 2),
            nn.GELU(),
            nn.Conv2d(embed_dim // 2, embed_dim, patch_size, patch_size, bias=False),
            nn.BatchNorm2d(embed_dim),
        )
        
        # Positional embedding
        self.pos_embed = nn.Parameter(torch.zeros(1, self.num_patches, embed_dim))
        nn.init.trunc_normal_(self.pos_embed, std=0.02)
        
        # Mamba blocks
        self.blocks = nn.ModuleList([
            MambaBlock(embed_dim, d_state, expand)
            for _ in range(depth)
        ])
        
        # Global average pooling
        self.pool = nn.AdaptiveAvgPool1d(1)
        
        # Feature projection for ZSL
        self.feature_proj = nn.Sequential(
            nn.Linear(embed_dim, embed_dim // 2),
            nn.GELU(),
            nn.Dropout(0.1),
            nn.Linear(embed_dim // 2, embed_dim // 2)
        )
        
    def forward(self, x):
        B = x.shape[0]
        
        # Patch embedding
        x = self.patch_embed(x)  # (B, embed_dim, H', W')
        x = x.flatten(2).transpose(1, 2)  # (B, num_patches, embed_dim)
        
        # Add positional embedding
        x = x + self.pos_embed
        
        # Apply Mamba blocks
        for block in self.blocks:
            x = block(x)
        
        # Global pooling and feature projection
        x = x.transpose(1, 2)  # (B, embed_dim, num_patches)
        x = self.pool(x).squeeze(-1)  # (B, embed_dim)
        
        features = self.feature_proj(x)  # (B, embed_dim//2)
        return features

# Efficient Zero-Shot Learning Classifier
class EfficientZSLClassifier(nn.Module):
    def __init__(self, feature_dim, num_classes, semantic_dim=64):
        super().__init__()
        self.num_classes = num_classes
        self.semantic_dim = semantic_dim
        
        # Learnable semantic embeddings (compact)
        self.semantic_embeddings = nn.Parameter(torch.randn(num_classes, semantic_dim))
        nn.init.xavier_uniform_(self.semantic_embeddings)
        
        # Feature to semantic mapping
        self.semantic_proj = nn.Sequential(
            nn.Linear(feature_dim, semantic_dim),
            nn.LayerNorm(semantic_dim)
        )
        
        # Temperature parameter for scaled dot-product
        self.temperature = nn.Parameter(torch.ones(1) * 10.0)
        
    def forward(self, features):
        # Project features to semantic space
        semantic_features = self.semantic_proj(features)  # (B, semantic_dim)
        
        # Normalize embeddings
        semantic_features = F.normalize(semantic_features, dim=-1)
        semantic_embeddings = F.normalize(self.semantic_embeddings, dim=-1)
        
        # Compute similarity scores
        logits = torch.matmul(semantic_features, semantic_embeddings.t())  # (B, num_classes)
        logits = logits * self.temperature
        
        return logits

# Model instantiation
embed_dim = 96
model = CompactVisionMamba(embed_dim=embed_dim).to(device)
zsl_classifier = EfficientZSLClassifier(
    feature_dim=embed_dim // 2, 
    num_classes=num_classes, 
    semantic_dim=64
).to(device)

# Parameter counting function
def count_parameters(model):
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    return total_params, trainable_params

# Count parameters
vision_total, vision_trainable = count_parameters(model)
zsl_total, zsl_trainable = count_parameters(zsl_classifier)
total_params = vision_total + zsl_total

print(f"\n=== Model Parameters ===")
print(f"Vision Mamba: {vision_total:,} total, {vision_trainable:,} trainable")
print(f"ZSL Classifier: {zsl_total:,} total, {zsl_trainable:,} trainable")
print(f"Total Parameters: {total_params:,}")
print(f"Target: < 1M parameters âœ“" if total_params < 1_000_000 else f"Target: < 1M parameters âœ—")

# Optimizer with different learning rates
optimizer = optim.AdamW([
    {'params': model.parameters(), 'lr': 0.001, 'weight_decay': 0.0001},
    {'params': zsl_classifier.parameters(), 'lr': 0.001, 'weight_decay': 0.0001}
])

# Learning rate scheduler
scheduler = optim.lr_scheduler.CosineAnnealingWarmRestarts(optimizer, T_0=10, T_mult=2)

# Mixed precision training
scaler = torch.cuda.amp.GradScaler() if device.type == 'cuda' else None

# Training tracking
train_losses, val_losses = [], []
train_accs, val_accs = [], []
best_val_acc = 0.0

# Time analysis variables
epoch_times = []
inference_times = []

def evaluate():
    model.eval()
    zsl_classifier.eval()
    val_loss = 0.0
    correct = 0
    total = 0
    all_preds = []
    all_labels = []
    all_probs = []
    
    # Time inference
    inference_start = time.time()
    
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device, non_blocking=True), labels.to(device, non_blocking=True)
            
            with torch.cuda.amp.autocast() if scaler else torch.no_grad():
                features = model(images)
                logits = zsl_classifier(features)
                loss = F.cross_entropy(logits, labels)
            
            val_loss += loss.item()
            probs = F.softmax(logits, dim=1)
            preds = torch.argmax(probs, dim=1)
            
            correct += (preds == labels).sum().item()
            total += labels.size(0)
            
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            all_probs.extend(probs.cpu().numpy())
    
    inference_time = time.time() - inference_start
    accuracy = 100.0 * correct / total
    return val_loss / len(val_loader), accuracy, all_preds, all_labels, all_probs, inference_time

# Training loop
print(f"\n=== Starting Training ===")
num_epochs = 60
training_start_time = time.time()

for epoch in range(1, num_epochs + 1):
    epoch_start_time = time.time()
    
    model.train()
    zsl_classifier.train()
    train_loss = 0.0
    correct = 0
    total = 0
    
    pbar = tqdm(train_loader, desc=f"Epoch {epoch:2d}/{num_epochs}")
    for images, labels in pbar:
        images, labels = images.to(device, non_blocking=True), labels.to(device, non_blocking=True)
        
        optimizer.zero_grad()
        
        if scaler:
            with torch.cuda.amp.autocast():
                features = model(images)
                logits = zsl_classifier(features)
                loss = F.cross_entropy(logits, labels)
            
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
        else:
            features = model(images)
            logits = zsl_classifier(features)
            loss = F.cross_entropy(logits, labels)
            loss.backward()
            optimizer.step()
        
        train_loss += loss.item()
        preds = torch.argmax(logits, dim=1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)
        
        # Update progress bar
        train_acc = 100.0 * correct / total
        pbar.set_postfix({
            'Loss': f'{train_loss/len(pbar):.4f}',
            'Acc': f'{train_acc:.2f}%'
        })
    
    scheduler.step()
    
    # Validation
    val_loss, val_acc, val_preds, val_labels, val_probs, inf_time = evaluate()
    
    # Track metrics
    train_losses.append(train_loss / len(train_loader))
    val_losses.append(val_loss)
    train_accs.append(train_acc)
    val_accs.append(val_acc)
    
    # Time tracking
    epoch_time = time.time() - epoch_start_time
    epoch_times.append(epoch_time)
    inference_times.append(inf_time)
    
    # Save best model
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'zsl_state_dict': zsl_classifier.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'val_acc': val_acc,
        }, 'best_model.pth')
    
    print(f"Epoch {epoch:2d}/{num_epochs} - Train Loss: {train_losses[-1]:.4f}, "
          f"Acc: {train_acc:.2f}% | Val Loss: {val_loss:.4f}, Acc: {val_acc:.2f}% "
          f"| Best: {best_val_acc:.2f}% | Time: {epoch_time:.2f}s")

total_training_time = time.time() - training_start_time

# Final evaluation on best model
checkpoint = torch.load('best_model.pth')
model.load_state_dict(checkpoint['model_state_dict'])
zsl_classifier.load_state_dict(checkpoint['zsl_state_dict'])

final_val_loss, final_val_acc, final_preds, final_labels, final_probs, final_inf_time = evaluate()

# === CLASSIFICATION REPORT ===
print(f"\n" + "="*50)
print(f"CLASSIFICATION REPORT")
print(f"="*50)
class_report = classification_report(
    final_labels, 
    final_preds, 
    target_names=train_dataset.classes,
    digits=4
)
print(class_report)

# === TIME ANALYSIS REPORT ===
print(f"\n" + "="*50)
print(f"TIME ANALYSIS REPORT")
print(f"="*50)
print(f"Total Training Time: {total_training_time:.2f} seconds ({total_training_time/60:.2f} minutes)")
print(f"Average Epoch Time: {np.mean(epoch_times):.2f} Â± {np.std(epoch_times):.2f} seconds")
print(f"Fastest Epoch: {np.min(epoch_times):.2f} seconds")
print(f"Slowest Epoch: {np.max(epoch_times):.2f} seconds")
print(f"Average Inference Time: {np.mean(inference_times):.4f} Â± {np.std(inference_times):.4f} seconds")
print(f"Fastest Inference: {np.min(inference_times):.4f} seconds")
print(f"Slowest Inference: {np.max(inference_times):.4f} seconds")
print(f"Training Speed: {len(train_dataset)/(np.mean(epoch_times)):.2f} samples/second")
print(f"Inference Speed: {len(val_dataset)/np.mean(inference_times):.2f} samples/second")

# === ABLATION STUDY REPORT ===
print(f"\n" + "="*50)
print(f"ABLATION STUDY REPORT")
print(f"="*50)

def ablation_test(model_variant, description, variant_embed_dim=None):
    """Simplified ablation test - runs a few epochs to estimate performance"""
    print(f"\nTesting: {description}")
    
    # Quick training for 5 epochs
    test_model = model_variant.to(device)
    
    # Use the correct feature dimension for the variant
    if variant_embed_dim is None:
        variant_embed_dim = embed_dim
    
    test_classifier = EfficientZSLClassifier(
        feature_dim=variant_embed_dim // 2, 
        num_classes=num_classes, 
        semantic_dim=64
    ).to(device)
    
    test_optimizer = optim.AdamW([
        {'params': test_model.parameters(), 'lr': 0.001},
        {'params': test_classifier.parameters(), 'lr': 0.001}
    ])
    
    test_acc = 0
    start_time = time.time()
    
    for epoch in range(3):  # Quick test
        test_model.train()
        test_classifier.train()
        
        for i, (images, labels) in enumerate(train_loader):
            if i > 10:  # Only few batches for quick test
                break
                
            images, labels = images.to(device), labels.to(device)
            test_optimizer.zero_grad()
            
            features = test_model(images)
            logits = test_classifier(features)
            loss = F.cross_entropy(logits, labels)
            loss.backward()
            test_optimizer.step()
    
    # Quick evaluation
    test_model.eval()
    test_classifier.eval()
    correct = 0
    total = 0
    
    with torch.no_grad():
        for i, (images, labels) in enumerate(val_loader):
            if i > 5:  # Quick eval
                break
            images, labels = images.to(device), labels.to(device)
            features = test_model(images)
            logits = test_classifier(features)
            preds = torch.argmax(logits, dim=1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)
    
    test_time = time.time() - start_time
    test_acc = 100.0 * correct / total if total > 0 else 0
    test_params = sum(p.numel() for p in test_model.parameters()) + sum(p.numel() for p in test_classifier.parameters())
    
    return test_acc, test_params, test_time

# Ablation variants
print("1. Baseline (Current Model)")
print(f"   Accuracy: {final_val_acc:.2f}%")
print(f"   Parameters: {total_params:,}")
print(f"   Inference Time: {final_inf_time:.4f}s")

# Variant 1: Reduced embedding dimension
model_v1 = CompactVisionMamba(embed_dim=64)  # Reduced from 96
acc_v1, params_v1, time_v1 = ablation_test(model_v1, "Reduced Embedding Dimension (64)", variant_embed_dim=64)
print(f"   Quick Test Accuracy: {acc_v1:.2f}%")
print(f"   Parameters: {params_v1:,} ({(params_v1/total_params-1)*100:+.1f}%)")
print(f"   Time: {time_v1:.2f}s")

# Variant 2: Fewer Mamba blocks
model_v2 = CompactVisionMamba(embed_dim=96, depth=4)  # Reduced from 6
acc_v2, params_v2, time_v2 = ablation_test(model_v2, "Fewer Mamba Blocks (4)", variant_embed_dim=96)
print(f"   Quick Test Accuracy: {acc_v2:.2f}%")
print(f"   Parameters: {params_v2:,} ({(params_v2/total_params-1)*100:+.1f}%)")
print(f"   Time: {time_v2:.2f}s")

# Variant 3: Smaller state dimension
model_v3 = CompactVisionMamba(embed_dim=96, d_state=8)  # Reduced from 16
acc_v3, params_v3, time_v3 = ablation_test(model_v3, "Smaller State Dimension (8)", variant_embed_dim=96)
print(f"   Quick Test Accuracy: {acc_v3:.2f}%")
print(f"   Parameters: {params_v3:,} ({(params_v3/total_params-1)*100:+.1f}%)")
print(f"   Time: {time_v3:.2f}s")

print(f"\nAblation Summary:")
print(f"- Reducing embedding dim: {acc_v1-final_val_acc:+.2f}% accuracy, {(params_v1/total_params-1)*100:+.1f}% parameters")
print(f"- Fewer Mamba blocks: {acc_v2-final_val_acc:+.2f}% accuracy, {(params_v2/total_params-1)*100:+.1f}% parameters")
print(f"- Smaller state dim: {acc_v3-final_val_acc:+.2f}% accuracy, {(params_v3/total_params-1)*100:+.1f}% parameters")

print(f"\n=== Final Results ===")
print(f"Best Validation Accuracy: {final_val_acc:.2f}%")
print(f"Total Parameters: {total_params:,}")
print(f"Parameters < 1M: {'âœ“' if total_params < 1_000_000 else 'âœ—'}")
print(f"Accuracy â‰¥ 99%: {'âœ“' if final_val_acc >= 99.0 else 'âœ—'}")

# Plotting results
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# Loss curves
ax1.plot(train_losses, label='Train Loss', color='blue')
ax1.plot(val_losses, label='Val Loss', color='red')
ax1.set_title('Loss Over Epochs')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.legend()
ax1.grid(True)

# Accuracy curves
ax2.plot(train_accs, label='Train Accuracy', color='blue')
ax2.plot(val_accs, label='Val Accuracy', color='red')
ax2.axhline(y=99, color='green', linestyle='--', label='Target 99%')
ax2.set_title('Accuracy Over Epochs')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy (%)')
ax2.legend()
ax2.grid(True)

# Confusion Matrix
cm = confusion_matrix(final_labels, final_preds)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax3,
            xticklabels=train_dataset.classes, yticklabels=train_dataset.classes)
ax3.set_title('Confusion Matrix')
ax3.set_xlabel('Predicted')
ax3.set_ylabel('True')

# ROC Curves
final_labels_np = np.array(final_labels)
final_probs_np = np.array(final_probs)
y_true = label_binarize(final_labels_np, classes=list(range(num_classes)))

colors = plt.cm.Set1(np.linspace(0, 1, num_classes))
for i, color in enumerate(colors):
    fpr, tpr, _ = roc_curve(y_true[:, i], final_probs_np[:, i])
    roc_auc = auc(fpr, tpr)
    ax4.plot(fpr, tpr, color=color, lw=2, 
             label=f'{train_dataset.classes[i]} (AUC = {roc_auc:.3f})')

ax4.plot([0, 1], [0, 1], 'k--', lw=2)
ax4.set_xlim([0.0, 1.0])
ax4.set_ylim([0.0, 1.05])
ax4.set_xlabel('False Positive Rate')
ax4.set_ylabel('True Positive Rate')
ax4.set_title('ROC Curves - Multi-class')
ax4.legend(loc="lower right", fontsize='small')

plt.tight_layout()
plt.show()

print(f"\nðŸš€ Training completed! Best accuracy: {best_val_acc:.2f}%")
```
